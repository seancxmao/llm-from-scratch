```python
import torch
from model import GPT2Model
from config import GPT2_CONFIG_124M
from weights import load_weights
from tokenizer import GPT2Tokenizer
from generation import generate

gpt = GPT2Model(GPT2_CONFIG_124M)
load_weights(gpt)
gpt.eval();

torch.manual_seed(123)
tokenizer = GPT2Tokenizer()
input_ids = tokenizer.encode("Every effort moves you")
output_ids = generate(
    model=gpt,
    idx=input_ids,
    max_new_tokens=25,
    context_size=GPT2_CONFIG_124M["context_length"],
    top_k=50,
    temperature=1.5
)
print("Output text:\n", tokenizer.decode(output_ids))
```
